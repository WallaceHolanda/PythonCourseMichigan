## Juntando Dataframes

```python
# Criando dois dataframes - funcionários e alunos
import pandas as pd

staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])

# Realizando a indexação por meio do nome
staff_df = staff_df.set_index('Name')

student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])


student_df = student_df.set_index('Name')

print(staff_df.head())
print(student_df.head())
```

```python
# Realizando a junção dos dataframes 
# Os dados foram agrupados a partir do nome

pd.merge(staff_df, student_df, how = 'outer', left_index = True, right_index = True)
```

```python
# Agrupando os dados da interseção - que são tanto alunos como funcionários

pd.merge(staff_df, student_df, how = 'inner', left_index = True, right_index = True)
```

```python
# Agrupando os dados que são funcionário, independe se são ou não alunos

pd.merge(staff_df, student_df, how = 'left', left_index = True, right_index = True)
```

```python
# Agrupando os dados que são alunos, independe se são ou não funcionários

pd.merge(staff_df, student_df, how = 'right', left_index = True, right_index = True)
```

```python
# Resetando os índices

staff_df = staff_df.reset_index()
student_df = student_df.reset_index()

# Usando o parâmetro 'on' - coletando todos que são alunos
pd.merge(staff_df, student_df, how='right', on='Name')
```

## Idioma do Pandas

```python
import pandas as pd
import numpy as np
import timeit

df = pd.read_csv('bases/census.csv')
df.head()
```

```python
# Difinir o estado e a cidade como um índice múltiplo
# Somente paras as tuplas de nível de resumo iguais a 50
# Por fim, uma coluna será renomeada

(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))
```

```python
# Realizando a mesma operação mas como funções individuais
df = df[df['SUMLEV']==50] 
df.set_index(['STNAME','CTYNAME'], inplace=True)
df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
# Verificando o desempenho da primeira abordagem

def first_approach():
    global df
    # And we'll just paste our code right here
    return (df.where(df['SUMLEV']==50)
             .dropna()
             .set_index(['STNAME','CTYNAME'])
             .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

df = pd.read_csv('bases/census.csv')
timeit.timeit(first_approach, number=10)
```

```python
# Verificando o desempenho da segunda abordagem

def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME','CTYNAME'], inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})

df = pd.read_csv('bases/census.csv')
timeit.timeit(second_approach, number=10)
```

```python
# Criando uma função de mapeamento

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})
```

```python
# Aplicando a função eixo das colunas
df.apply(min_max, axis='columns').head()
```

```python
# Outra forma de realizar o mapeamento

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    
    # Cria um novo campo para o valor máximo
    row['max'] = np.max(data)
    
    # Cria um novo campo para o valor mínimo
    row['min'] = np.min(data)
    return row

# Now just apply the function across the dataframe
df.apply(min_max, axis='columns')
```

```python
# Criando uma lambda para retornar o valor máximo

rows = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013','POPESTIMATE2014', 
        'POPESTIMATE2015']
# Now we'll just apply this across the dataframe with a lambda
df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

```python
# Dividindo a localidade em regiões

def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
```

```python
# Criando uma nova coluna para a região
# Aplicando a lambda para cada instância
df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```

```python
# Verificando o resultado da aplicação
df[['STNAME','state_region']].head()
```

## Grupos

```python
import pandas as pd
import numpy as np

# Lendo o dataframe e seperando os dados
df = pd.read_csv('bases/census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 3

# Verificando a média populacional para cada cidade

for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    
    print('Counties in state ' + state + 
          ' have an average population of ' + str(avg))
```

```python
%%timeit -n 3

# Utilizando a abordagem de groupby

for group, frame in df.groupby('STNAME'):
    
    avg = np.average(frame['CENSUS2010POP'])
    
    print('Counties in state ' + group + 
          ' have an average population of ' + str(avg))
```

```python
# Criando uma função para verificar o primeiro caractere
# Definindo o grupo a partir do caractere 0, 1 ou 2
df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

# frame - as tuplas / group - o grupo (0, 1 ou 2)
for group, frame in df.groupby(set_batch_number):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
```

```python
# Carregando o dataframe

df=pd.read_csv("bases/listings.csv")
df.head()
```

```python
# Definindo índices múltiplos
df=df.set_index(["cancellation_policy","review_scores_value"])

# Verificando os possíveis grupos
for group, frame in df.groupby(level=(0,1)):
    print(group)
```

```python
def grouping_fun(item):
    # Verificando o "review_scores_value" do índice
    if item[1] == 10.0:
        return (item[0],"10.0")
    else:
        return (item[0],"not 10.0")

for group, frame in df.groupby(by=grouping_fun):
    print(group)
```

```python
df.head()
```

```python
# Agregação

# Resetando os índices
df = df.reset_index()

# Agrupando pela política de cancelamento e obtendo a média
df.groupby("cancellation_policy").agg({"review_scores_value":np.average})
```

```python
# Agg retorna um valor por coluna - nesse caso, a média
df.groupby("cancellation_policy").agg({"review_scores_value":np.nanmean})
```

## Transformação

```python
# Retorna um objeto do mesmo tamanho do grupo

# definindo as colunas de interesse
cols=['cancellation_policy','review_scores_value']

# Transformando e armazenando no dataframe
transform_df=df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
# Renomeando o nome da coluna criada
transform_df.rename({'review_scores_value':'mean_review_scores'}, axis='columns', inplace=True)

# Juntando o df original com o transformado
df=df.merge(transform_df, left_index=True, right_index=True)
df.head()
```

```python
# Criando uma nova coluna mean_diff
# Armazenará a diferença da avaliação obtida em relação a média
df['mean_diff']=np.absolute(df['review_scores_value']-df['mean_review_scores'])
df['mean_diff'].head()
```

```python
# Filtragem
# Filtrando as instâncias com média superior a 9.2
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
```

```python
# Carregando os novos dados
df=pd.read_csv("bases/listings.csv")

df=df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
# Aplicando funções arbitrária aos grupos
# Por meio do apply será possível obter a média e calcular o desvio
# em uma só operação, ao contrário da abordagem anterior

def calc_mean_review_scores(group):
    avg = np.nanmean(group["review_scores_value"])
    
    group["review_scores_mean"]=np.abs(avg-group["review_scores_value"])
    return group

df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```

## Tabela de Pivô

```python
import pandas as pd
import numpy as np

df = pd.read_csv('bases/cwurData.csv')
df.head()
```

```python
# Criando um novo atributo para categorizar as universidades

def create_category(ranking):
    if (ranking >= 1) & (ranking <= 100):
        return "First Tier Top Unversity"
    elif (ranking >= 101) & (ranking <= 200):
        return "Second Tier Top Unversity"
    elif (ranking >= 201) & (ranking <= 300):
        return "Third Tier Top Unversity"
    return "Other Top Unversity"

# Aplicando a função por universidade usando o lambda
df['Rank_Level'] = df['world_rank'].apply(lambda x: create_category(x))
df.head()
```

```python
# Agrupando as médias dos resultados por país

df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean]).head()
```

```python
# Aplicando a média e o valor máximo - lista de funções
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max]).head()
```

```python
# Agora coleta o maior de todos e também a média de todas as categorias

df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max], 
               margins=True).head()
```

```python
# Criando um novo df a partir da tabela anterior
new_df = df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max], 
               margins=True)

# Verificando os índices e colunas
print(new_df.index)
print(new_df.columns)
```
